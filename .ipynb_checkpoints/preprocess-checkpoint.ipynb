# Data Preprocessing

```python
print(data_json.info())
```

```python
# Lower casing and remove punctuations
data_json['reviewText'] = data_json['reviewText'].str.lower()
data_json['reviewText'] = data_json['reviewText'].str.translate(str.maketrans('', '', string.punctuation))

# Tokenize
lemmatizer = WordNetLemmatizer()
data_json['reviewText'] = data_json['reviewText'].apply(lambda x: word_tokenize(str(x)))

# Remove stop words
spacy_model = spacy.load('en_core_web_sm')
stopwords_spacy = spacy_model.Defaults.stop_words
data_json['reviewText'] = data_json['reviewText'].apply(lambda x: [word for word in x if word not in stopwords_spacy])

# POS Tagging
data_json['POStag'] = data_json['reviewText'].apply(lambda x: pos_tag(x))

# Lemmatization
data_json['reviewText'] = data_json['reviewText'].apply(lambda x: [lemmatizer.lemmatize(token) for token in x])

# Join back as a lemmatized sentence
data_json['reviewText'] = data_json['reviewText'].apply(lambda x: ' '.join(x))
```

# Sentiment Analysis using VaderSentiment 

# From Vader documentation:

Compound Score: computed by summing the valence scores of each word in the lexicon, adjusted according to the rules, and then normalized to be between -1 (most extreme negative) and +1 (most extreme positive). This is the most useful metric if you want a single unidimensional measure of sentiment for a given sentence. Calling it a 'normalized, weighted composite score' is accurate.

The pos, neu, and neg scores are ratios for proportions of text that fall in each category (so these should all add up to be 1... or close to it with float operation). These are the most useful metrics if you want to analyze the context & presentation of how sentiment is conveyed or embedded in rhetoric for a given sentence.


```python
analyzer = SentimentIntensityAnalyzer()
data_json['vadersentiment'] = data_json['reviewText'].apply(lambda x: analyzer.polarity_scores(str(x)))
data_json['vaderNeg'] = data_json['vadersentiment'].apply(lambda x: x['neg'])
data_json['vaderNeu'] = data_json['vadersentiment'].apply(lambda x: x['neu'])
data_json['vaderPos'] = data_json['vadersentiment'].apply(lambda x: x['pos'])
data_json['vaderComp'] = data_json['vadersentiment'].apply(lambda x: x['compound'])

data_json.drop(['vadersentiment'], axis=1, inplace=True)
```

```python
def vaderCompSen(compoundscore = 0):
    if compoundscore > posthreshold:
        return 'Positive'
    elif compoundscore < negthreshold:
        return 'Negative'
    else: 
        return 'Neutral'

data_json['vaderCompSen'] = data_json['vaderComp'].apply(lambda x: vaderCompSen(x))
```

```python
plt.figure(figsize=(10, 6))
order = ['Negative', 'Neutral', 'Positive']
sns.countplot(data=data_json, x='vaderCompSen', order = order)
plt.title('Distribution of Review Sentiments based on Vader Compound')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
def vaderMaxSen(neg, neu, pos):
    if pos == max(neg, neu, pos):
        return 'Positive'
    elif neg == max(neg, neu, pos):
        return 'Negative'
    else: 
        return 'Neutral'

data_json['vaderMaxSen'] = data_json.apply(lambda row: vaderMaxSen(row['vaderNeg'], row['vaderNeu'], row['vaderPos']), axis=1)
```

```python
plt.figure(figsize=(10, 6))
order = ['Negative', 'Neutral', 'Positive']
sns.countplot(data=data_json, x='vaderMaxSen', order = order)
plt.title('Distribution of Review Sentiments based on Maximum Vader Score')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
data_json[['vaderNeg', 'vaderNeu', 'vaderPos']]
```

```python
accuracy = accuracy_score(data_json['vaderMaxSen'], data_json['sentiment'])
f1 = f1_score(data_json['vaderMaxSen'], data_json['sentiment'], average='weighted')

# Print the results
print(f"Using the maximum between the Positive, Neutral, and Negative Score, the Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}")

accuracy = accuracy_score(data_json['vaderCompSen'], data_json['sentiment'])
f1 = f1_score(data_json['vaderCompSen'], data_json['sentiment'], average='weighted')

# Print the results
print(f"Using the Compound Score, the Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}")
```

# Sentiment Analysis using TextBlob 

# From TextBlob documentation:

The sentiment property returns a namedtuple of the form Sentiment(polarity, subjectivity). The polarity score is a float within the range [-1.0, 1.0]. The subjectivity is a float within the range [0.0, 1.0] where 0.0 is very objective and 1.0 is very subjective.

```python
data_json['textblobsentiment'] = data_json['reviewText'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
data_json['textblobsub'] = data_json['reviewText'].apply(lambda x: TextBlob(str(x)).sentiment.subjectivity)
data_json['textblobsubjectivity'] = data_json['textblobsub'].apply(lambda x: 'Subjective' if x > 0.5 else 'Objective')
```

As the scoring of TextBlob is also from [-1.0, 1.0], which is the same as the compound score in Vader, the same threshold used in vaderCompSen() is used.

```python
data_json['textblobSen'] = data_json['textblobsentiment'].apply(lambda x: vaderCompSen(x))
```

```python
plt.figure(figsize=(10, 6))
order = ['Negative', 'Neutral', 'Positive']
sns.countplot(data=data_json, x='textblobSen', order = order)
plt.title('Distribution of Review Sentiments based on TextBlob Compound')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data_json, x='textblobsubjectivity')
plt.title('Distribution of Review Subjectivity based on TextBlob')
plt.xlabel('Subjectivity')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
accuracy = accuracy_score(data_json['textblobSen'], data_json['sentiment'])
f1 = f1_score(data_json['textblobSen'], data_json['sentiment'], average='weighted')

print(f"Using the TextBlob Score, the Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}")
```

# Sentiment Analysis using SentiWordNet

```python
from nltk.corpus import wordnet as wn
from nltk.corpus import sentiwordnet as swn
from nltk.stem import WordNetLemmatizer

def penn_to_wn(tag):
    if tag.startswith('J'):
        return wn.ADJ
    elif tag.startswith('N'):
        return wn.NOUN
    elif tag.startswith('R'):
        return wn.ADV
    elif tag.startswith('V'):
        return wn.VERB
    return None

# Initialize sentiment and tokens_count
sentiment = 0.0
tokens_count = 0

def calculate_sentiment(pos_tags):
    sentiment = 0.0
    tokens_count = 0

    for word, tag in pos_tags:
        wn_tag = penn_to_wn(tag)
        if wn_tag not in (wn.NOUN, wn.ADJ, wn.ADV):
            continue
        
        lemmatizer = WordNetLemmatizer()
        lemma = lemmatizer.lemmatize(word, pos=wn_tag)
        if not lemma:
            continue

        # Get synsets
        synsets = list(wn.synsets(lemma, pos=wn_tag))
        if not synsets:
            continue

        synset = synsets[0]
        swn_synset = swn.senti_synset(synset.name())

        sentiment += swn_synset.pos_score() - swn_synset.neg_score()
        tokens_count += 1

    # Check if tokens_count is not zero before dividing
    if tokens_count != 0:
        sentiment /= tokens_count

    return sentiment

data_json['SWNSentimentScore'] = data_json['POStag'].apply(calculate_sentiment)
```

```python
def SWNSentiment(sentiment_score):
    if sentiment_score > posthreshold:
        return 'Positive'
    elif sentiment_score < negthreshold:
        return 'Negative'
    else:
        return 'Neutral'

# Apply the classification function to each row in 'Sentiment' column
data_json['SWNSentiment'] = data_json['SWNSentimentScore'].apply(SWNSentiment)
```

```python
plt.figure(figsize=(10, 6))
order = ['Negative', 'Neutral', 'Positive']
sns.countplot(data=data_json, x='SWNSentiment', order = order)
plt.title('Distribution of Review Sentiments based on SWNSentiment')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
accuracy = accuracy_score(data_json['SWNSentiment'], data_json['sentiment'])
f1 = f1_score(data_json['SWNSentiment'], data_json['sentiment'], average='weighted')

print(f"Using the SWN Score, the Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}")
```
